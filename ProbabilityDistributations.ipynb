```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
#setting seabon color style
sns.set(color_codes=True)
#settings for seaborn plot sizes
sns.set(rc = {'figure.figsize':(10,7)})
```

```python
from scipy.stats import uniform
```

generating 10 random uniform distributions from 0 to 10

```python
n = 10000
a = 0
b = 10
data_uniform = uniform.rvs(size = n,loc = a,scale = b)
```

```python
#plotting the uniform distribution using distplot

ax = sns.distplot(data_uniform,bins = 10,color= 'red')
ax.set(xlabel = 'uniform',ylabel = 'freq')
```

```python
ax = sns.distplot(data_uniform,bins = 10,color = 'red',kde = False,
                 hist_kws = {'linewidth':15})
```

Generating random numbers for bernoullo distribution

```python
from scipy.stats import bernoulli
```

```python
data_bern = bernoulli.rvs(size = 10000,p = .2) # p = success probability
```

```python
ax = sns.distplot(data_bern,color = 'black',
                  hist_kws = {"linewidth":15,'alpha':1})
ax.set(xlabel='Bernoulli', ylabel='Frequency')
```
